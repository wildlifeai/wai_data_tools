```python
import wandb
import pathlib
from wai_data_tools import actions
from wai_data_tools.utils import setup_logging
```

```python
setup_logging.setup_logging(logging_dir=None, logging_config_file=None)
```

```python
raw_dataset_dir = pathlib.Path("")
proc_dataset_dir = raw_dataset_dir.parent / "processed"
```

```python
actions.filter_empty_videos(src=raw_dataset_dir, dest=proc_dataset_dir, dry_run=False, threshold=225)
```

```python
# Save dataset to WANDB server
wandb.init(project="kupe-skinks")
artifact = wandb.Artifact(name="dataset", type="dataset")
artifact.add_dir(proc_dataset_dir, name="dataset")
wandb.finish()
```

```python
# Load dataset from WANDB
dataset_dir = pathlib.Path("/Users/dava/Documents/kupe-skinks/dataset")
dataset_dir.mkdir(exist_ok=True, parents=True)

run = wandb.init(project="kupe-skinks")
artifact = run.use_artifact("dataset:latest")
artifact.download(dataset_dir)
wandb.finish()
```

```python
actions.create_dataset(dataset_name="kupe-skinks", data_dir=dataset_dir)
```

```python
actions.show_dataset("kupe-skinks")
```
